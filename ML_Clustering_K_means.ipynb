```python
import pandas as pd
from sklearn import datasets
from sklearn.cluster import KMeans
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

**Évalue ton modèle en utilisant la méthode Elbow, entre 2 et 10 clusters, quel nombre cluster semble optimal ?**

```python
mylist=[]
for i in range (2,11):
  modelKM2 = KMeans(n_clusters=i,random_state=3)
  modelKM2.fit(df_data)
  mylist.append(modelKM2.inertia_)
```

```python
plt.plot(range(2,11),mylist,'-*')
plt.ylabel('inertia')
plt.xlabel('k clusters')
plt.title('Elbow method');
```

**Évalue ton modèle en utilisant la méthode Silhouette, entre 2 et 10 clusters, quel nombre de clusters semble optimal ? Et toi, visuellement, si tu ne savais pas qu'il y avait 3 espèces, tu en aurais trouvé combien ?**

```python
from sklearn.metrics  import silhouette_score
mylist=[]
for k in range(2,11):
    modelKM2 = KMeans(n_clusters=k, random_state=3)
    modelKM2.fit(X)
    mylist.append(silhouette_score(df_data, modelKM2.labels_))
```

```python
plt.plot(range(2,11),mylist,'-*')
plt.ylabel('silhouette')
plt.xlabel('k clusters')
plt.title('silhouette method');
```
