```python
inertia=[]

for k in range(1,20):
  modelKM=KMeans(n_clusters=k).fit(X)
  inertia.append(modelKM.inertia_)
inertia
```

```python
plt.plot(range(1,20),inertia,'-+')
plt.title('Elbow method')
plt.xlabel('No of clusters')
plt.ylabel('inertia')
plt.show()
```

***nombre cluster semble optimal  3***

```python
# Ici, nous faisons une boucle pour afficher le score Silhouette

# pour un nombre de clusters entre 2 et 8

from sklearn.metrics  import silhouette_score

for k in range(2,9):

    modelKM = KMeans(n_clusters=k, random_state=3)

    modelKM.fit(X)

    print(silhouette_score(X, modelKM.labels_))
```

**nombre de clusters semble optima 2 (methode silhouette)**

```python
silhouette=[]
for k in range(2,9):
    modelKM = KMeans(n_clusters=k, random_state=3).fit(X)
    silhouette.append(silhouette_score(X, modelKM.labels_))
silhouette
```

```python
plt.plot(range(2,9),silhouette,'-+')
plt.title('Silhouette method')
plt.xlabel('No of clusters')
plt.ylabel('silhouette')
plt.show()
```
