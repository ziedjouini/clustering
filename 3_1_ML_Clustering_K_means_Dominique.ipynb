```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
# 1 Dans scikit-learn, importe le dataset Iris, c'est un grand classique de la data.
# Il contient 3 espèces de fleurs (des iris) suivant leur longueur et largeur de pétale et de sépale.
df_iris = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
```

```python
#2 Entraîne ton modèle pour qu’il trouve 3 clusters à partir des 4 colonnes numériques.
from sklearn.cluster import KMeans
X = df_iris[['sepal_length','sepal_width','petal_length','petal_width']]
y = df_iris.species
modelKM = KMeans(n_clusters=3, random_state=0)
modelKM.fit(X)
```

```python
# 6 Évalue ton modèle en utilisant la méthode Elbow, entre 2 et 10 clusters,
# quel nombre de cluster semble optimal ? -> 4

inertias = []
K = range(2, 11)

for k in K:
    # Building and fitting the model
    kmeanModel = KMeans(n_clusters=k, n_init='auto').fit(X)
    kmeanModel.fit(X)
    inertias.append(kmeanModel.inertia_)

print(inertias)

plt.plot(K, inertias, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Inertia')
plt.title('The Elbow Method using Inertia')
plt.show()
```

```python
# Évalue ton modèle en utilisant la méthode Silhouette, entre 2 et 10 clusters,
# quel nombre de clusters semble optimal ? -> 2
# Et toi, visuellement, si tu ne savais pas qu'il y avait 3 espèces, tu en aurais trouvé combien ? -> 2
from sklearn.metrics import silhouette_score
silhouettes = []
K = range(2, 11)

for k in K:
    # Building and fitting the model
    modelKM = KMeans(n_clusters=k, random_state=3, n_init='auto')
    modelKM.fit(X)
    silhouettes.append(silhouette_score(X, modelKM.labels_))

print(silhouettes)

plt.plot(K, silhouettes, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('silhouettes')
plt.title('silhouettes')
plt.show()

```
